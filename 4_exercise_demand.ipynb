```python
df_US = df[df['job_country']=='United States'].copy()
```

```python
df_US
```

```python
df_US['job_month'] = df['job_posted_date'].dt.month
```

```python
df_US
```

```python
df_US['job_month_var'] = df_US['job_posted_date'].dt.month_name()
```

```python
df_US['job_counts'] = df_US.groupby('job_month')['job_month'].value_counts()
```

```python
df_US
```

```python
df_final = pd.pivot_table(
    df_US,
    index = ['job_month','job_month_var'],
    columns = 'job_title_short',
    values = 'job_posted_date',
    aggfunc = 'count'
)
```

```python
df_final.sort_index(level='job_month')
```

```python
df_final = df_final.reset_index().set_index('job_month_var')
```

```python
df_final
```

```python
print(df_final.index.names)
print(df_final.columns)
```

```python
df_final = df_final.reset_index()  # Moves all index levels to columns
# Now drop as a column:
df_final.drop('job_month', axis=1, inplace=True)
```
